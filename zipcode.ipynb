```python
p_hood = pd.read_csv("HomePrice_Neighborhood.csv")
p_hood.shape
p_hood.head()
```

```python
atlanta = p_hood[p_hood['City']=='Atlanta']
atlanta.shape
atlanta.head()
```

```python
atlantadrop = atlanta.drop(['RegionID','City','State','Metro','CountyName','SizeRank'], axis=1)
```

```python
T_patlanta = atlantadrop.set_index('RegionName').T
T_patlanta.index = pd.to_datetime(T_patlanta.index)
T_patlanta.shape
T_patlanta.head()
```

```python
T_patlanta = T_patlanta.fillna(0)
T_patlanta = T_patlanta.astype(float)
T_patlanta['Total']=T_patlanta.iloc[:,0:58].sum(axis=1)
# T_patlanta['Average']=T_patlanta[T_patlanta.iloc[:,0:58]!=0].mean(axis=1)
T_patlanta['Average']=T_patlanta.iloc[:,0:58].mean(axis=1)
T_patlanta.head(1)
```

```python
T_patlanta['Normalized'] = T_patlanta['Average'] / T_patlanta['Average'].loc[0,]
```

```python
T_pzip['Normalized'] = T_pzip['Average'] / T_pzip['Average'].loc[0,]
```
