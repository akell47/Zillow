```python
p_hood = pd.read_csv("HomePrice_Neighborhood.csv")
p_hood.shape
p_hood.head()
```

```python
atlanta = p_hood[p_hood['City']=='Atlanta']
atlanta.shape
atlanta.head()
```

```python
atlantadrop = atlanta.drop(['RegionID','City','State','Metro','CountyName','SizeRank'], axis=1)
```

```python
T_patlanta = atlantadrop.set_index('RegionName').T
T_patlanta.index = pd.to_datetime(T_patlanta.index)
T_patlanta.shape
T_patlanta.head()
```

```python
T_patlanta = T_patlanta.fillna(0)
T_patlanta = T_patlanta.astype(float)
T_patlanta['Total']=T_patlanta.iloc[:,0:58].sum(axis=1)
T_patlanta['Average']=T_patlanta[T_patlanta.iloc[:,0:58]!=0].mean(axis=1)
T_patlanta.head(1)
```

```python
T_patlanta[('Average')].resample('M').mean().plot(figsize=(20,6))
```

```python
T_patlanta[('Ansley Park')].resample('M').mean().plot(figsize=(20,6))
```

```python
patlanta = T_patlanta.drop(['Total'], axis=1)
```

```python
neighborhoodList = list(patlanta.columns.values)
patlanta[neighborhoodList].resample('M').mean().plot(figsize=(20,100))
```

```python
df = patlanta

avgP = go.Scatter(
    x=df.index,
    y=df[('Average')],
    name = "Atlanta Neighborhoods",
    line = dict(color = '#17BECF'),
    opacity = 0.8)

# trace_low = go.Scatter(
#     x=df.Date,
#     y=df['AAPL.Low'],
#     name = "AAPL Low",
#     line = dict(color = '#7F7F7F'),
#     opacity = 0.8)

data = [avgP]

layout = dict(
    title='Average Atlanta Home Values',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
#                 dict(count=1,
#                      label='1m',
#                      step='month',
#                      stepmode='backward'),
                dict(count=6,
                     label='6m',
                     step='month',
                     stepmode='backward'),
                dict(step='all')
            ])
        ),
        rangeslider=dict(),
        type='date'
    )
)

fig = dict(data=data, layout=layout)
py.iplot(fig, filename = "HomePriceAtlanta")
```
